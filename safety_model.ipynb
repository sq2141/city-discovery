```python
import pandas as pd
import numpy as np
```

```python
# load cities df
cities = pd.read_csv('cities_text_processed_df.csv', index_col = 0)
# safearound.com data
safety = pd.read_csv('cities_safety_ratings_manual.csv')
# Global Peace Index
peace = pd.read_csv('countries_peace_index.csv')
```

```python
# Add city safety rating
cities2 = pd.merge(cities,safety[['City','Country','Safety index']], how = 'left', on = ['City','Country'])
```

```python
# Add country peace score
peace['Country'] = peace['Country'].apply(lambda x: x.replace(u'\xa0', u'')); # Clean up weird character
cities3 = pd.merge(cities2,peace, how = 'left', on = 'Country')
```

```python
safety_df = cities3[['Country','Understand','Stay safe','Lat','Lon','2018 score','Safety index']]
```
