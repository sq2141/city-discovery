```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import ast
    
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from textblob import TextBlob


import warnings
with warnings.catch_warnings():
    warnings.filterwarnings("ignore",category=DeprecationWarning)
    import nltk
    from nltk.tokenize import RegexpTokenizer
    from nltk.stem.wordnet import WordNetLemmatizer
    from nltk.corpus import stopwords, words
    from nltk import FreqDist
```

```python
# load cities df
cities = pd.read_csv('cities_text_processed_df.csv', index_col = 0)
# If reloaded from .csv file, convert string to list
if type(cities['Stay safe_tokens'][0])==str:
    cities['Stay safe_tokens'] = cities['Stay safe_tokens'].apply(ast.literal_eval)


# safearound.com data
safety = pd.read_csv('cities_safety_ratings_manual.csv')
# Global Peace Index
peace = pd.read_csv('countries_peace_index.csv')
```

```python
# Add city safety rating
cities2 = pd.merge(cities,safety[['City','Country','Safety index']], how = 'left', on = ['City','Country'])
```

```python
# Add country peace score
peace['Country'] = peace['Country'].apply(lambda x: x.replace(u'\xa0', u'')); # Clean up weird character
cities3 = pd.merge(cities2,peace, how = 'left', on = 'Country')
cities3['my_index'] = cities3.index
```

```python
safety_df = cities3[['Country','Stay safe','Lat','Lon','2018 score','Safety index']]
```

# Predict safety

```python
safety_df.head()
```

```python
# Features (exclude WikiTravel text for now)
safety_noWT = safety_df[['Country','Lat','Lon','2018 score','Safety index']]
safety_noWT_filled = safety_noWT[['Country','Lat','Lon','2018 score']].fillna(safety_noWT[['Country','Lat','Lon','2018 score']].mean())
safety_noWT_filled['Safety index'] = safety_noWT['Safety index']
safety_noWT_filled.head()
```

```python
# one hot encoding
encoded_data = pd.get_dummies(safety_noWT_filled)
encoded_data.insert(0,'my_index',cities3.index)
encoded_data.head()
```

```python
encoded_data_with_values = encoded_data[encoded_data['Safety index'].notnull()]
encoded_data_without_values = encoded_data[encoded_data['Safety index'].isnull()]

X = encoded_data_with_values.drop(['my_index','Safety index'], axis=1)
Y = encoded_data_with_values['Safety index']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.25, random_state = 42)
```

## Linear regression

```python
lm = LinearRegression()
lm.fit(X_train,Y_train)
# Predict on test set
pred = lm.predict(X_test)
plt.scatter(Y_test,pred)
```

```python
# Calculate mean square error and root mean square error
mse_cost = np.mean((pred-Y_test)**2)
rmse_cost = np.sqrt(mse_cost)

print('The RMSE of cost is ' + str(round(rmse_cost,2)))
```

## Random Forest

```python
rfm = RandomForestRegressor()
rfm.fit(X_train,Y_train)
# Predict on test set
pred = rfm.predict(X_test)
plt.scatter(Y_test,pred)
```

```python
# Calculate mean square error and root mean square error
mse_cost = np.mean((pred-Y_test)**2)
rmse_cost = np.sqrt(mse_cost)

print('The RMSE of cost is ' + str(round(rmse_cost,2)))
```

# Regression including text data

```python
safety = safety_df[['Stay safe','Country','Lat','Lon','2018 score','Safety index']]
```

```python
safety['Stay safe'].fillna('', inplace=True)
safety.head()
```

```python
safety['Safety length']=safety['Stay safe'].apply(lambda x: len(x))
safety['Sentiment']=safety['Stay safe'].apply(lambda x: TextBlob(x).sentiment.polarity)
safety.drop('Stay safe', axis=1, inplace=True)
```

```python
def make_nan(sentiment):
    if (sentiment<=0.000001) & (sentiment>=-0.000001):
        return np.nan
    else:
        return sentiment

safety['Sentiment'] = safety['Sentiment'].apply(make_nan)
```

```python
safety.head()
```

```python
plt.scatter(safety['Sentiment'],safety['Safety index'])
```

```python
# Features
safety_filled = safety[['Country','Lat','Lon','2018 score','Sentiment']].fillna(safety[['Country','Lat','Lon','2018 score','Sentiment']].mean())
safety_filled['Safety index'] = safety['Safety index']
safety_filled['Safety length'] = safety['Safety length']
safety_filled.head()
```

```python
# one hot encoding
encoded_data = pd.get_dummies(safety_filled)
encoded_data.insert(0,'my_index',cities3.index)
encoded_data.head()
```

```python
encoded_data_with_values = encoded_data[encoded_data['Safety index'].notnull()]
encoded_data_without_values = encoded_data[encoded_data['Safety index'].isnull()]

X = encoded_data_with_values.drop(['my_index','Safety index'], axis=1)
Y = encoded_data_with_values['Safety index']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.25, random_state = 42)
```

## Linear regression

```python
lm = LinearRegression()
lm.fit(X_train,Y_train)
# Predict on test set
pred = lm.predict(X_test)
plt.scatter(Y_test,pred)
```

```python
# Calculate mean square error and root mean square error
mse_cost = np.mean((pred-Y_test)**2)
rmse_cost = np.sqrt(mse_cost)

print('The RMSE of cost is ' + str(round(rmse_cost,2)))
```

## Random Forest

```python
rfm = RandomForestRegressor()
rfm.fit(X_train,Y_train)
# Predict on test set
pred = rfm.predict(X_test)
plt.scatter(Y_test,pred)
```

```python
# Calculate mean square error and root mean square error
mse_cost = np.mean((pred-Y_test)**2)
rmse_cost = np.sqrt(mse_cost)

print('The RMSE of cost is ' + str(round(rmse_cost,2)))
```
